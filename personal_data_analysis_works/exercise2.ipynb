### 10. [10 pts] Subset the original dataset to the years 1995-1997. Group the data by team name and then summarize the average number of points per team. Arrange from most to least points. <a class="anchor" id="bullet10"></a>

```python
#subset the data from year 1995 - 1997:
nba_95_97 = nba[(nba["Year"] >= 1995.0) & (sort_by_score["Year"] <= 1997.0)]
#group by team name and year, and calculate avg number of points:
nba_groupby_y = nba_95_97.groupby(["Tm", "Year"])["PTS"].agg([np.mean])
#group by team name only, and calculate avg number of points:
nba_groupby = nba_95_97.groupby(["Tm"])["PTS"].agg([np.mean])
#sort the groupby scores:
nba_groupby_sort = nba_groupby.sort_values(by = "mean", ascending = True).round(1)
nba_groupby_sort
```

### 11. [10 pts] Reshape this data in the previous question into a wide format. <a class="anchor" id="bullet11"></a>Create a wide dataset that keeps year in a single column, but spreads team names to multiple individual columns with each column delineating points per team in 1995-1997. 

```python
#reset the data index so it could be reshape using pivot:
nba_groupby_sort = nba_groupby_y.reset_index()
#reshape the data to wide format:
reshaped_nba = nba_groupby_sort.pivot(index="Year", columns="Tm", values="mean")
reshaped_nba
```

### 12. [10 pts] Now return the data to a long (tidy) <a class="anchor" id="bullet12"></a>format by moving teams back into a single column and points in a single column. Hint: data should include three columns: one for year, one for teams, and one for points. 

```python
#reshape the function so it could be melted:
reshaped_nba_index = reshaped_nba.reset_index()
#melt the function:
back_to_long_nba = pd.melt(reshaped_nba_index, id_vars="Year")
back_to_long_nba
```
