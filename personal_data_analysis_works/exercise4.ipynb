#### 7. [15 pts] <a class="anchor" id="bullet7"></a>This question does not use the dataset used in the previous HW questions. Give a function that accepts unlimited arguments, checks and counts how many data types are in those arguments. Finally return the total in a list. This list should count the following object types, in this order:

```python
def count_datatypes(*args):
    # create a dictionary to hold the count of each data type
    counts = {
        "int": 0,
        "str": 0,
        "bool": 0,
        "list": 0,
        "tuple": 0,
        "dict": 0
    }
    
    # loop through each argument
    for arg in args:
        # get the type of the argument
        arg_type = type(arg).__name__
        
        # increment the count for the corresponding data type
        if arg_type == "int":
            counts["int"] += 1
        elif arg_type == "str":
            counts["str"] += 1
        elif arg_type == "bool":
            counts["bool"] += 1
        elif arg_type == "list":
            counts["list"] += 1
        elif arg_type == "tuple":
            counts["tuple"] += 1
        elif arg_type == "dict":
            counts["dict"] += 1
    
    # return the counts as a list:
    return [counts["int"], counts["str"], counts["bool"], counts["list"], counts["tuple"], counts["dict"]]
#try this example:
count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1)
```
